Implementering af Dayes algoritme:

```python
from Bio import Phylo

def count_duplicates(duplicate): 
    final_list = []
    num_dup = 0
    for num in duplicate: 
        if num not in final_list: 
            final_list.append(num)
        else:
            num_dup += 1
    return num_dup 

def RF_distance(filename1, filename2):
    # Read in trees in Newick format and reroot with seq1 ad root
    tree1 = Phylo.read(filename1, "newick")
    tree2 = Phylo.read(filename2, "newick")
    tree1.root_with_outgroup("seq1")
    tree2.root_with_outgroup("seq1")
    n = len(tree1.get_terminals())
    
    # Make ORIG and DF table for tree1
    ORIG = [0]*n
    DF = [0]*n
    for node in range(0,n):
        name = str(tree1.get_terminals()[node])
        idxname = int(name[3:])-1
        DF[idxname] = node
        ORIG[node] = idxname
    
    #Find clades in tree1
    term_names = [term.name for term in tree1.get_terminals()]
    term_names.sort()
    clades1 = []
    for clade in tree1.get_nonterminals():
        clade_term_names = [term.name for term in clade.get_terminals()]
        clades1.append(clade_term_names)
    for clade in range(len(clades1)):
        for seq in range(len(clades1[clade])):
            seq_name = clades1[clade][seq]
            idxname = int(seq_name[3:])-1
            clades1[clade][seq] = DF[idxname]
    
    #Find clades in tree2 and rename them according to DF name of tree1
    clades2=[]
    for clade in tree2.get_nonterminals():
        clade_term_names = [term.name for term in clade.get_terminals()]
        clades2.append(clade_term_names)
    for clade in range(len(clades2)):
        for seq in range(len(clades2[clade])):
            seq_name = clades2[clade][seq]
            idxname = int(seq_name[3:])-1
            clades2[clade][seq] = DF[idxname]
        clades2[clade].sort()
    
    #Find total number of clades and subtract duplicates times 2
    clades_total = clades1+clades2
    num_clades = len(clades_total)
    num_dup = count_duplicates(clades_total)
    return num_clades-2*num_dup



print(RF_distance("tree1.new", "tree2.new"))
    
```
